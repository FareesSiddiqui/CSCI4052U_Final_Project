```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image_path = "image.png"
img = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Get the dimensions of the image
height, width = gray.shape

# Assume 8x8 chessboard, calculate the size of each square
square_height = height // 8
square_width = width // 8

# List to store cropped images
cropped_images = []

# Loop over the 8x8 grid and crop each square
for row in range(8):
    row_images = []
    for col in range(8):
        x_start = col * square_width
        y_start = row * square_height
        x_end = (col + 1) * square_width
        y_end = (row + 1) * square_height
        cropped_square = img[y_start:y_end, x_start:x_end]
        row_images.append(cropped_square)
    cropped_images.append(row_images)

# Reverse the order of rows to align with the standard chessboard orientation
correctly_oriented_images = cropped_images[::-1]

# Display a sample cropped image from a1 again (which should now be the bottom-left of the image)
plt.imshow(cv2.cvtColor(correctly_oriented_images[7][0], cv2.COLOR_BGR2RGB))
plt.title("Sample cropped square from a8")
plt.axis('off')
plt.show()

```

```python
# Invert the image vertically to make the white pieces appear at the bottom
correct_orientation_grid = correctly_oriented_images[::-1]

# Set up a plot with 8x8 subplots (corrected orientation)
fig, axs = plt.subplots(8, 8, figsize=(8, 8))

# Remove the axes for a cleaner look
for row in range(8):
    for col in range(8):
        axs[row, col].imshow(cv2.cvtColor(correct_orientation_grid[row][col], cv2.COLOR_BGR2RGB))
        axs[row, col].axis('off')

# Display the entire 8x8 grid with the correct orientation (white at the bottom)
plt.tight_layout()
plt.show()
```
